```python
import matplotlib
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# read the data file and print the first lines
filename = "data/stocked.csv"
df = pd.read_csv(filename, header=0)
print(df.head())
```

```python
plot_index = range(1, len(df['proximity'])+1)
print(plot_index)
```

```python
plt.rcParams["figure.figsize"] = (15,11)
plt.plot(plot_index, df['proximity'], color='green', label="proximity", linestyle="solid", marker=',')
plt.plot(plot_index, df['shade'], color='blue', label="shade", linestyle="solid", marker=',')
plt.plot(plot_index, df['brightness'], color='red', label="brightness", linestyle="solid", marker=',')
plt.title("STOCKED ACTION")
plt.xlabel("Sample #")
plt.ylabel("Sample value")
plt.legend()
plt.show()
```
